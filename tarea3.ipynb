# Filtro de Blur

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Replace 'input_image.jpg' with the path to your input image
input_image_path = 'Covid19-dataset/train/Covid/01.jpeg'
covid_image = cv2.imread(input_image_path, cv2.IMREAD_GRAYSCALE)
input_image_path = 'Covid19-dataset/train/Normal/01.jpeg'
normal_image = cv2.imread(input_image_path, cv2.IMREAD_GRAYSCALE)
input_image_path = 'Covid19-dataset/train/Viral Pneumonia/01.jpeg'
pneumonia_image = cv2.imread(input_image_path, cv2.IMREAD_GRAYSCALE)

# Gaussian kernel size (must be an odd number)
kernel_size = 5

# Create the Gaussian kernel
kernel = np.ones((kernel_size, kernel_size), np.float32) / (kernel_size ** 2)

# Perform the convolution
filtered_covid_image = cv2.filter2D(covid_image, -1, kernel)
filtered_normal_image = cv2.filter2D(normal_image, -1, kernel)
filtered_pneumonia_image = cv2.filter2D(pneumonia_image, -1, kernel)


# ploteo de las imagenes en una celda de 3 filas y 2 columnas
plt.figure(figsize=(10, 12))

plt.subplot(3, 2, 1)
plt.imshow(covid_image, cmap='gray')
plt.title('Original Covid Image')
plt.axis('off')

plt.subplot(3, 2, 2)
plt.imshow(filtered_covid_image, cmap='gray')
plt.title('Filtered Image')
plt.axis('off')

plt.subplot(3, 2, 3)
plt.imshow(normal_image, cmap='gray')
plt.title('Original Normal Image')
plt.axis('off')

plt.subplot(3, 2, 4)
plt.imshow(filtered_normal_image, cmap='gray')
plt.title('Filtered Image')
plt.axis('off')

plt.subplot(3, 2, 5)
plt.imshow(pneumonia_image, cmap='gray')
plt.title('Original Pneumonia Image')
plt.axis('off')

plt.subplot(3, 2, 6)
plt.imshow(filtered_pneumonia_image, cmap='gray')
plt.title('Filtered Image')
plt.axis('off')

plt.tight_layout()
plt.show()

```

# Filtro de Sobel Edge Detection

```python
# Aplicar el operador Sobel en direccin X
sobel_x = cv2.Sobel(covid_image, cv2.CV_64F, 1, 0, ksize=3)

# Aplicar el operador Sobel en direccin Y
sobel_y = cv2.Sobel(covid_image, cv2.CV_64F, 0, 1, ksize=3)

# Combinar las imgenes resultantes de Sobel en direccin X e Y
filtered_covid_image_2 = cv2.sqrt(cv2.addWeighted(cv2.pow(sobel_x, 2.0), 1.0, cv2.pow(sobel_y, 2.0), 1.0, 0.0))

# Se repite el proceso para las otra 2 imgenes
sobel_x = cv2.Sobel(normal_image, cv2.CV_64F, 1, 0, ksize=3)
sobel_y = cv2.Sobel(normal_image, cv2.CV_64F, 0, 1, ksize=3)
filtered_normal_image_2 = cv2.sqrt(cv2.addWeighted(cv2.pow(sobel_x, 2.0), 1.0, cv2.pow(sobel_y, 2.0), 1.0, 0.0))

sobel_x = cv2.Sobel(pneumonia_image, cv2.CV_64F, 1, 0, ksize=3)
sobel_y = cv2.Sobel(pneumonia_image, cv2.CV_64F, 0, 1, ksize=3)
filtered_pneumonia_image_2 = cv2.sqrt(cv2.addWeighted(cv2.pow(sobel_x, 2.0), 1.0, cv2.pow(sobel_y, 2.0), 1.0, 0.0))

# ploteo
plt.figure(figsize=(10, 12))

plt.subplot(3, 2, 1)
plt.imshow(covid_image, cmap='gray')
plt.title('Original Covid Image')
plt.axis('off')

plt.subplot(3, 2, 2)
plt.imshow(filtered_covid_image_2, cmap='gray')
plt.title('Filtered Image')
plt.axis('off')

plt.subplot(3, 2, 3)
plt.imshow(normal_image, cmap='gray')
plt.title('Original Normal Image')
plt.axis('off')

plt.subplot(3, 2, 4)
plt.imshow(filtered_normal_image_2, cmap='gray')
plt.title('Filtered Image')
plt.axis('off')

plt.subplot(3, 2, 5)
plt.imshow(pneumonia_image, cmap='gray')
plt.title('Original Pneumonia Image')
plt.axis('off')

plt.subplot(3, 2, 6)
plt.imshow(filtered_pneumonia_image_2, cmap='gray')
plt.title('Filtered Image')
plt.axis('off')

plt.tight_layout()
plt.show()
```

# Canny Edge Detection

```python
# Aplicar el filtro de deteccin de bordes Canny
bordes = cv2.Canny(covid_image, 100, 200)  # Los valores 100 y 200 son umbral inferior y superior respectivamente

bordes2 = cv2.Canny(normal_image, 100, 200)

bordes3 = cv2.Canny(pneumonia_image, 100, 200)


# ploteo
plt.figure(figsize=(10, 12))

plt.subplot(3, 2, 1)
plt.imshow(covid_image, cmap='gray')
plt.title('Original Covid Image')
plt.axis('off')

plt.subplot(3, 2, 2)
plt.imshow(bordes, cmap='gray')
plt.title('Filtered Image')
plt.axis('off')

plt.subplot(3, 2, 3)
plt.imshow(normal_image, cmap='gray')
plt.title('Original Normal Image')
plt.axis('off')

plt.subplot(3, 2, 4)
plt.imshow(bordes2, cmap='gray')
plt.title('Filtered Image')
plt.axis('off')

plt.subplot(3, 2, 5)
plt.imshow(pneumonia_image, cmap='gray')
plt.title('Original Pneumonia Image')
plt.axis('off')

plt.subplot(3, 2, 6)
plt.imshow(bordes3, cmap='gray')
plt.title('Filtered Image')
plt.axis('off')

plt.tight_layout()
plt.show()
```

# Filtro aplicado Anisotropic Diffusion 

```python
# Aplicar el filtro de difusin anisotrpica
filtro_difusion = cv2.edgePreservingFilter(covid_image, flags=1, sigma_s=60, sigma_r=0.4)

filtro_difusion2 = cv2.edgePreservingFilter(normal_image, flags=1, sigma_s=60, sigma_r=0.4)

filtro_difusion3 = cv2.edgePreservingFilter(pneumonia_image, flags=1, sigma_s=60, sigma_r=0.4)


# ploteo
plt.figure(figsize=(10, 12))

plt.subplot(3, 2, 1)
plt.imshow(covid_image, cmap='gray')
plt.title('Original Covid Image')
plt.axis('off')

plt.subplot(3, 2, 2)
plt.imshow(filtro_difusion, cmap='gray')
plt.title('Filtered Image')
plt.axis('off')

plt.subplot(3, 2, 3)
plt.imshow(filtro_difusion2, cmap='gray')
plt.title('Original Normal Image')
plt.axis('off')

plt.subplot(3, 2, 4)
plt.imshow(filtro_difusion2, cmap='gray')
plt.title('Filtered Image')
plt.axis('off')

plt.subplot(3, 2, 5)
plt.imshow(pneumonia_image, cmap='gray')
plt.title('Original Pneumonia Image')
plt.axis('off')

plt.subplot(3, 2, 6)
plt.imshow(filtro_difusion3, cmap='gray')
plt.title('Filtered Image')
plt.axis('off')

plt.tight_layout()
plt.show()
```
